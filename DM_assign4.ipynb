```python
type(df)
```

```python
from mlxtend.frequent_patterns import apriori,association_rules 
```

```python
freq_item=apriori(df,min_support=0.3,use_colnames=True)
freq_item
```

```python
asso_rule=association_rules(freq_item,metric="confidence",min_threshold=0.8)
asso_rule
asso_rule = asso_rule.sort_values(['confidence'], ascending =[False])
asso_rule 
```
