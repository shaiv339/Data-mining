```python
#Group the data by number of bedrooms and get the maximum and minimum prices for each group
df3=df2.groupby('beds')['price'].max()
print(df3)
```

```python
df3=df2.groupby('beds')['price'].min()
print(df3)
```

```python
df3=df2.groupby('beds')['price'].agg(['min','max'])
print(df3)
```

3) On average, which is a costly city in respect of the prices of flats?

```python
df3=df2.groupby('city')['price'].mean()
most_expensive_city=df3.idxmax()
print("The most expensive city is:",most_expensive_city)
```

4) Which flats are more costly? Residential or condo? What is the average price?

```python
result=df2.groupby('type')['price'].max()
result2=df2.groupby('type')['price'].mean()
if(result[0]>result[2]):
  print("Condo has the highest flat prices with average price of flats:",result2[0])
else:
  print("Residential has the highest flat prices with average price of flats:",result2[2])
```
