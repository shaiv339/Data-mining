Find out frequent itemset and generate rules out of **it**

```
%matplotlib inline
import numpy as np
import pandas as pd
```

```
df1=pd.read_csv('/content/Cosmetics - Cosmetics.csv')
df1.head()
```

```
df = df1.replace({
                'Yes': True,
                'No': False
            })
df
```

```
from mlxtend.frequent_patterns import apriori, association_rules
frq_items=apriori(df, min_support=0.2,use_colnames=True)
frq_items
```

```
frq_items['length']=frq_items['itemsets'].apply(lambda x: len(x))
frq_items
```

```
rules = association_rules(frq_items, metric ="confidence", min_threshold = 0.7) 
rules = rules.sort_values(['confidence'], ascending =[False]) 
rules
```
